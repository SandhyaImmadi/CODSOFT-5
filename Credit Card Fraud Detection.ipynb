```python
#seperating the data for analysis
legit = ccd[ccd.Class == 0]
fraud = ccd[ccd.Class == 1]
```

```python
#Under Sampling- the idea is to build a sample dataset containing similar distribution of legit and fraudulent transactions
#total no.of fraudulent transactions - 492
```

```python
legit_sample = legit.sample(n=492)
```

```python
#concatenating legit_sample and fraud dataframes
new_ccd = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_ccd.head()
```

```python
new_ccd.tail()
```

```python
new_ccd['Class'].value_counts()
```

```python
new_ccd.groupby('Class').mean()
```

```python
#Above the nature of the dataset has been maintained 
```

```python
#splitting the data into training and testing data
X = new_ccd.drop(columns='Class' , axis=1)
Y = new_ccd['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, stratify=Y, random_state=2)
#stratify used for even distribution of legal and fraud transactions in test and train dataset
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model training
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#model evaluation
X_train_prediction=model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on testing data: ',testing_data_accuracy)
```

```python
from sklearn import metrics
print(metrics.classification_report(X_test_prediction,Y_test))
```
